Pandas : an opensource lib for data analysis and preprocessing that leverages power and speed of numpy(fast)
pandas has two types of data structures:
1. Series : 1D array with indexes,stores a single coln or row of data
2. Dataframe : Tabular data structure with columns of different types of data

```python
import numpy as np
import pandas as pd
```

DataFrame is a 2D table which has rows and columns and has indexes for faster indexing

```python
dict1 = {
    "name":['komal','piyush','rohan'],
    "class":['15','10','12'],
    "skill":['faster learning','mechanical','rider']
}
df = pd.DataFrame(dict1)
```

But when we make a copy of the dataframe, then change done in copy is not seen in the dataframe

```python
newdf1 = dataf.copy()
```

```python
newdf1[0][0] = 0
```

```python
dataf
```

```python
dataf.columns = list("ABCDE")
```

When i write dataf[0][0] = 78 this chaining of [0][0] returns view or copy as intermediate is not defined so when we want to set some value we use loc function

```python
dataf.loc[0,'A'] = 89
```

```python
dataf
```

```python
dataf.loc[0,0] = 90
```

```python
dataf
```

As we can see we got an extra column as 0 so we want to drop it

```python
dataf = dataf.drop(0,axis=1)
```

```python
dataf.loc[[1,2],['C','D']] # returns a copy or new dataframe who has only 1,2 rows and C,D columns
```

```python
dataf.loc[:,['C','D']]# returns all rows but C,D columns only
```

```python
dataf.loc[[1,2],:] # returns only 1,2 rows but all columns
```

```python
dataf.loc[:] # returns all rows and all columns
```

```python
dataf.loc[(dataf['A']<0.9)]
```
